### Pars ABSA 

```python
#Pars-ABSA all
pars_absa_comments = []
pars_absa_labels = [] 
pars_absa_targets = []
with open('sentiment_datasets\Pars-ABSA\Pars-ABSA_all.txt', 'r', encoding='utf-8') as ff:
    lines = ff.readlines()
    for i, line in enumerate(lines):
        if(i%3 == 0):
            pars_absa_comments.append(line.rstrip('\n'))
            continue
        if(i%3 == 2):
            pars_absa_labels.append(int(line.split('\n')[0]))
            continue
        if(i%3 == 1):
            pars_absa_targets.append(line)
```

```python
pars_absa_dataset = pd.DataFrame(list(zip(pars_absa_comments, pars_absa_labels)), columns=['comment', 'label'])
pars_absa_dataset
```

```python
pars_absa_dataset.label.value_counts()
```

### Emo Pars

```python
#persian_emotion_detection
persian_emotion_dataset = pd.read_csv('emotions_datasets\\Persian-Emotion-Detection\\dataset.csv')
persian_emotion_dataset
```

```python
import matplotlib.pyplot as plt   
emotion_labels = ['Anger', 'Fear', 'Happiness', 'Hatred', 'Sadness','Wonder']
yes_label_counts = [0 for e in emotion_labels]

for _, row in persian_emotion_dataset.iterrows():
    for e_id in range(len(emotion_labels)):
        e = emotion_labels[e_id]
        if row[e]>2:
            yes_label_counts[e_id] += 1

list(zip(emotion_labels, yes_label_counts))
```
